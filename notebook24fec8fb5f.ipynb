```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
dataset= pd.read_csv('../input/lt-vehicle-loan-default-prediction/train.csv')
dataset_test = pd.read_csv('../input/lt-vehicle-loan-default-prediction/test.csv')
dataset = dataset.drop(['Date.of.Birth','DisbursalDate','AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH','PERFORM_CNS.SCORE.DESCRIPTION'], axis = 1)
dataset_test = dataset_test.drop(['Date.of.Birth','DisbursalDate','AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH','PERFORM_CNS.SCORE.DESCRIPTION'], axis = 1)
dataset =dataset.dropna(axis = 0 , how = 'any')
dataset_test =dataset_test.dropna(axis = 0 , how = 'any')
```

```python
dataset.shape
```

```python
X= dataset.iloc[:,:35].values
X_test2=dataset_test.iloc[:,:].values
Y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encode', OneHotEncoder(),[8])],remainder = 'passthrough')
ct.fit(X)
X = ct.transform(X)
```

```python
X.shape
```

```python
ct.fit(X_test2)
X_test2 = ct.transform(X_test2)
```

```python
X_test2.shape
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size = 0.2 , random_state = 42)
```

```python
print(X_train)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)

print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(Y_pred,Y_test)
print(cm)
accuracy_score(Y_pred , Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_tree = KNeighborsClassifier(n_neighbors = 5)
classifier_tree.fit(X_train , Y_train)
```

```python
Y_pred_tree = classifier_tree.predict(X_test)
print(Y_pred_tree)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm_tree = confusion_matrix(Y_pred_tree,Y_test)
print(cm_tree)
accuracy_score(Y_pred_tree , Y_test)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters   =5 , init = 'k-means++',random_state = 42)
kmeans.fit(X_train, Y_train)
```

```python
Y_pred_kmeans = kmeans.predict(X_test)
print(Y_pred_kmeans)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm_kmeans = confusion_matrix(Y_pred_kmeans,Y_test)
print(cm_kmeans)
accuracy_score(Y_pred_kmeans , Y_test)
```
